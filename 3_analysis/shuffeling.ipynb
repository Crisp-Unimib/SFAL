```python
import pandas as pd

# percorso del file creato dallo script precedente
CSV_PATH = "final_sample_across_layers.csv"   # cambia se il nome è diverso
RANDOM_STATE = 28                             # per rendere lo shuffle riproducibile

# 1. carica il dataset finale
df = pd.read_csv(CSV_PATH)

# 2. shuffle intralayer
df_shuffled = (
    df.groupby("layer", group_keys=False)
      .apply(lambda g: g.sample(frac=1, random_state=RANDOM_STATE))
      .reset_index(drop=True)
)

# 3. (opzionale) salva o visualizza
#print(df_shuffled.head(10))
df_shuffled.to_csv("final_sample_across_layers_shuffled.csv", index=False)
```
